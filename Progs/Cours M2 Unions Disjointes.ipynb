# Chaines de bits de longueur n avec k 1

```python
def list_BSnk(n, k):
    if k > n: 
        return []
    if k == 0:
        return [n*[0]]
    return ([[0]+x for x in list_BSnk(n-1, k)] +
            [[1]+x for x in list_BSnk(n-1, k-1)])
```

```python
list_BSnk(5,2)
```

```python
def rank_BSnk(x):
    if not x: return 0
    if x[0] == 0:           # liste vide
        return rank_BSnk(x[1:])
    else:
        return binom(len(x)-1, sum(x)-1) + rank_BSnk(x[1:])

def unrank_BSnk(n, k, i):
    if n == 0:
        return []
    bn1k = binom(n-1, k)
    if i < bn1k:
        return [0]+unrank_BSnk(n-1, k, i)
    else:
        return [1]+unrank_BSnk(n-1, k-1, i-bn1k)
```

```python
rank_BSnk([0,1,0,1,0])
```

```python
list_BSnk(5,2)[4]
```

```python
unrank_BSnk(5,2,4)
```

# Retour sur les coefficients du binome

Recursion, memoisation et programmation dynamique

```python
def binom(n,k):
    if k > n:
        return 0
    if k == 0:
        return 1
    return binom(n-1, k) + binom(n-1, k-1)
```

```python
[[binom(n, k) for k in range(n+1)] for n in range(10)]
```

```python
binom(100, 50) # tres long...
```

Memoization en utilisant les decorateurs de Python/Sage:

```python
@cached_function
def binom(n, k):
    if k > n:
        return 0
    if k == 0:
        return 1
    return binom(n-1, k) + binom(n-1, k-1)
```

```python
binom(100, 50)
```

Memoization a la main:

```python
binomcache = {(0,0) : 1}
```

```python
def binom(n, k):
    if k > n: return 0
    try:
        return binomcache[n,k]
    except KeyError:
        if k == 0: 
            res = 1
        else:
            res = binom(n-1, k) + binom(n-1, k-1)
        binomcache[n,k] = res
        return res
```

```python
binom(100, 50)
```

```python
binomcache
```

Programmation dynamique

```python
def binomrow(n):
    row = [1]
    for r in range(n):
        row = [1] + [row[i-1]+row[i] for i in range(1, r+1)] + [1]
    return row
```

```python
[binomrow(i) for i in range(10)]
```

```python
binomrow(100)[50]
```

# Les permutations

```python
def unrank_perm(l, r):
    if not l:
        if r == 0:
            return []
        else:
            raise IndexError
    reccard = factorial(len(l)-1)
    ifirst = r // reccard
    irec = r % reccard
    l = l[:]
    first = l[ifirst]
    del l[ifirst]
    return [first] + unrank_perm(l, irec)
```

```python
unrank_perm([1,2,3,4,5], 12)
```

```python
N = 4
[unrank_perm(range(N), i) for i in range(factorial(N))]
```

# Les multipermutations

```python
def multinom(l):
    return factorial(sum(l)) / prod(factorial(i) for i in l)
```

```python
multinom((3,1,0,2))
```

```python
def unrank_multi(l, r):
    if sum(l) == 0:
        if r == 0:
            return []
        else:
            raise IndexError
    for i in range(len(l)):
        if l[i] > 0:
            l[i] -= 1 
            reccard = multinom(l)
            if r < reccard:
                return [i] + unrank_multi(l, r)
            l[i] += 1
            r -= reccard
    raise IndexError
```

```python
unrank_multi([3,1,0,2], 45)
```

```python
MS = [3,1,1] 
[unrank_multi(MS[:], i) for i in range(multinom(MS))]
```
